```python
import nibabel as nib
import numpy as np
from nibabel import Nifti1Image
```

```python
data = nib.load('/local_mount/space/megaera/2/users/data/sub-I55/sub-I55_sample-slice-005-023_desc-3daxis_enface.nii.gz').get_fdata()
```

```python
data.shape
```

```python
#compute the norm of the data along the last axis
norm_data = (data**2).sum(axis=-1)**0.5
```

```python
# norm the original data by the norm
normed_data = data / norm_data[..., None]
```

```python
normed_data =np.where(np.isfinite(normed_data),normed_data,0)
```
